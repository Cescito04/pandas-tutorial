> **Let's access to specific value like indexing, we have to function 'loc' and 'iloc'**


> **NB: when slicing with 'loc' the upper index is inclusive while with 'iloc' it's not inclusive**

```python
## loc allows us to filter our dataframes by rows and columns
coffee.loc[[0,1,5]]
coffee.loc[0:3]
coffee.loc[0:3, 'Day']
coffee.loc[2:6, ['Day', 'Units Sold']]
```

```python
## loc allows us to filter our dataframes by rows and columns only using index values
coffee.iloc[0:5, [0,2]]
```

> Let's modify some values by indexing

```python
coffee.loc[1, "Units Sold"] = 10
```

```python
coffee.head()
```

```python
coffee.sort_values('Units Sold')
```

```python
coffee.sort_values('Units Sold', ascending=False)
```

```python
coffee.sort_values(['Units Sold','Coffee Type'], ascending=False)
```

```python
coffee.sort_values(['Units Sold','Coffee Type'], ascending=[0,1])
```
