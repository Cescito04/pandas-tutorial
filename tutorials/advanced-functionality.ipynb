> Let's say we wanted to see how our revenue was doing comparated to the daye or something like aggregating things by week and then shifting to compare last week to this week, etc..

```python
coffee['yesterday-revenue'] = coffee['Revenue'].shift(1)
```

Let's see the percent changes


```python
coffee['pct-change'] = (coffee['Revenue'] - coffee['yesterday-revenue']) * 100
```
