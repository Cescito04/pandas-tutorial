> **So I think it might be interesting to start with filtering by height and weight**

```python
bios.loc[bios['height_cm'] > 215 ].head()
```

>let's try with specific columns

```python
bios.loc[bios['height_cm'] > 215, ['name','height_cm'] ].head()
```

> Without using the 'loc' method

```python
bios[bios['height_cm'] > 215]
```

```python
# With specific columns
bios[bios['height_cm'] > 215][['name','height_cm']].head()
```

> But what if we want to combine things, what if we wanted to have multiple conditions that we filter on

```python
bios.loc[(bios['height_cm'] > 215) & (bios['born_country'] == 'USA')]
```

>**We can start getting more more interesting with our conditionals by doing some specific filters based on string operations**

```python
bios[bios['name'].str.contains("Jean")]
```

> More cool with conditions 

```python
bios[bios['name'].str.contains("Jean|Patrick")]
```
