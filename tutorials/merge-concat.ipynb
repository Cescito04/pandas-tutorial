```python
import pandas as pd
import numpy as np
```

> **One thing that we might want to do is take this born country and we see just three letter-code, we might want to convert into the actual country name and that could be separate from where they compete** 

> I think we want to make it more explicit ways to compare where they're born abd where they competed and maybe just filter on data based on that

```python
nocs  = pd.read_csv('/Users/abdoulayebocoum/Desktop/data-ai-roadmap/pandas-tutorial/data/noc_regions.csv')
```

```python
nocs.head()
```

> We see that we have the threeL code and the region/country. So what we can do here is we can go ahead and merge the data

```python
bios_new = pd.merge(bios,nocs,left_on='born_country', right_on='NOC', how='left')
```

```python
bios_new.head()
```

> One thing to note is that because of the original df and the new df had this **NOC** column, it adds a suffix to bpth of them, but we can specify the suffix by doing .. :

```python
bios_new = pd.merge(bios,nocs,left_on='born_country', right_on='NOC', how='left', suffixes=['bios', 'nocdf'])
```

```python
bios_new.head()
```

```python
bios_new.rename(columns={'region': 'born_country_full'}, inplace=True)
```

```python
bios_new.head()
```

> Let's compare some columns

```python
bios_new[bios_new['NOCbios'] != bios_new['born_country_full']][['name','NOCbios','born_country_full']]
```
