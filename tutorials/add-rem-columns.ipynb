#### **Adding / Removing Columns**

```python
import pandas as pd
import numpy as np
```

```python
coffee = pd.read_csv('/Users/abdoulayebocoum/Desktop/data-ai-roadmap/pandas-tutorial/warmup-data/coffee.csv')
```

```python
coffee.head()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

> **Let's say we just wanted to add some sort of price column**

```python
coffee['price'] = 4.99
```

```python
coffee.head()
```

```python
coffee['new_price'] = np.where(coffee['Coffee Type']=='Espresso', 3.87, 5.99)
```

```python
coffee
```

> That looks cool but now we kind of have two prices, we don't want to prices so how do we drop that one column?

```python
coffee.drop(columns=['price'], inplace=True)
```

```python
coffee
```

> We could see it being very useful to have a column that's called revenue

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['new_price']
```

```python
coffee
```

> Another thing we might want to do is not 'new price' we might want to rename that column

```python
coffee.rename(columns={'new_price':'price'}, inplace=True)
```
