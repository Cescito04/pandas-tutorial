> Maybe we want to group by specific sport and then check what the average height or weight is, based on that sport
> Or maybe like we want sort by country and see what the average height of athletes coming from there

> But let's use the coffe dataframe

```python
coffee  = pd.read_csv('/Users/abdoulayebocoum/Desktop/Projects/pandas-tutorial/warmup-data/coffee.csv')
```

```python
coffee['Units Sold'] = coffee['Units Sold'].interpolate()
```

```python
coffee
```

> Maybe we wanted just to do some calculations on the coffee type, and maybe we wanted to grab the total number of units sold based the coffee type

```python
coffee.groupby('Coffee Type')['Units Sold'].sum()
```

```python
coffee.groupby('Coffee Type')['Units Sold'].mean()
```

> Maybe we wanted to calculate different things based on our data

```python
coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum',})
```

```python
coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum','Price': 'mean', 'Revenue': 'sum'})
```

```python
coffee.groupby(['Coffee Type','Day']).agg({'Units Sold': 'sum'})
```

> One way to do the same type of things but maybe is a litlle bit more easy to work with a longer term is to create a pivot our data
